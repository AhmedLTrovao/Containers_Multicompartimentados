**EXEMPLO 1**

```python
import gurobipy as gp
from gurobipy import GRB

model = gp.Model()
x = model.addVar(name="x", lb=0)
y = model.addVar(name="y", lb=0)

model.setObjective(x + y, GRB.MINIMIZE)
model.addConstr(x + 2*y >= 4, "c1")
model.addConstr(4*x + 3*y >= 12, "c2")

model.optimize()
for var in model.getVars():
    print(f"{var.VarName} = {var.X}")
print(f"Objective = {model.ObjVal}")
```

**EXEMPLO 2**

```python

# Cria o modelo
model = gp.Model("Mistura de Rações")

# Variáveis de decisão
x1 = model.addVar(name="x1", lb=0)
x2 = model.addVar(name="x2", lb=0)

# Função objetivo: minimizar custo
model.setObjective(3 * x1 + 2 * x2, GRB.MINIMIZE)

# Restrições
model.addConstr(2 * x1 + 1 * x2 >= 8, "Proteína")
model.addConstr(1 * x1 + 2 * x2 >= 6, "Fibra")

# Resolver
model.optimize()

# Resultados
print("\nSolução ótima:")
for v in model.getVars():
    print(f"{v.VarName} = {v.X:.2f}")
print(f"Custo total = R${model.ObjVal:.2f}")
```
